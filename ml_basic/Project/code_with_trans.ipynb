```python
# 범주형 데이터의 빈도수 분석
categorical_frequencies = emergency_data[categorical_columns].apply(pd.Series.value_counts)

# 범주형 데이터와 수치형 데이터를 분리
categorical_data = emergency_data.select_dtypes(include=['object'])
numerical_data = emergency_data.select_dtypes(include=['float64', 'int64'])

# 연속 변수와 이산 변수 구분
# 수치형 데이터 중 고유값의 개수가 많은 경우 연속 변수, 적은 경우 이산 변수로 간주
discrete_columns = [col for col in numerical_data.columns if numerical_data[col].nunique() < 20]
continuous_columns = [col for col in numerical_data.columns if numerical_data[col].nunique() >= 20]

discrete_data = numerical_data[discrete_columns]
continuous_data = numerical_data[continuous_columns]
```

```python
# 결측치 분포 분석
missing_values = emergency_data.isnull().sum().sort_values(ascending=False)
missing_percentage = (emergency_data.isnull().sum() / len(emergency_data) * 100).sort_values(ascending=False)
print(pd.concat([missing_values, missing_percentage], axis=1, keys=['Missing Values', 'Missing Percentage']))
```

```python
# 이상치 분석
# 수치형 데이터에 대한 IQR 기반 이상치 탐지
Q1 = numerical_data.quantile(0.25)
Q3 = numerical_data.quantile(0.75)
IQR = Q3 - Q1
outliers = ((numerical_data < (Q1 - 1.5 * IQR)) | (numerical_data > (Q3 + 1.5 * IQR))).sum().sort_values(ascending=False)
outliers_percentage = (outliers / len(numerical_data) * 100).sort_values(ascending=False)

print(pd.concat([outliers, outliers_percentage], axis=1, keys=['Outliers', 'Outliers Percentage']))
```

```python
from sklearn.utils.class_weight import compute_class_weight

# y_train에서 고유한 클래스 레이블 추출
unique_classes = np.unique(y_train)

# 클래스 가중치 계산
class_weights = compute_class_weight(
    class_weight='balanced',
    classes=unique_classes,  # 실제 y_train 데이터셋의 고유한 클래스 레이블을 사용합니다.
    y=y_train
)

# 클래스에 대한 가중치를 딕셔너리 형태로 매핑
# 클래스 레이블을 가중치와 매핑합니다.
class_weight_dict = {label: weight for label, weight in zip(unique_classes, class_weights)}

# 랜덤 포레스트 모델에 클래스 가중치 적용
rf_classifier = RandomForestClassifier(
    n_estimators=100, 
    random_state=42, 
    class_weight=class_weight_dict  # 수정된 클래스 가중치 딕셔너리를 사용합니다.
)

# 모델 학습
rf_classifier.fit(X_train, y_train)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from xgboost import XGBClassifier
from sklearn.preprocessing import LabelEncoder

# XGBoost 모델에 대한 하이퍼파라미터 검색 범위를 설정합니다.
xgb_params = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.8, 0.9, 1.0],
    'colsample_bytree': [0.8, 0.9, 1.0],
}

# RandomizedSearchCV를 설정합니다.
random_search = RandomizedSearchCV(
    estimator=XGBClassifier(),
    param_distributions=xgb_params,
    n_iter=10,  # 시도할 하이퍼파라미터 조합 수
    scoring='f1_macro',  # 불균형 데이터셋에 적합한 점수 함수
    cv=5,  # 5-fold 교차 검증
    random_state=42
)

# 레이블 인코더 초기화
label_encoder = LabelEncoder()

# y_train과 y_test에 레이블 인코딩 적용
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)

# 레이블 인코딩이 적용된 y_train_encoded로 모델 학습 및 RandomizedSearchCV 수행
random_search.fit(X_train, y_train_encoded)
```

```python
from lightgbm import LGBMClassifier

# LightGBM 하이퍼파라미터 검색 범위 설정
lgbm_params = {
    'n_estimators': [100, 200, 300],
    'num_leaves': [31, 50, 100],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.8, 0.9, 1.0],
    'colsample_bytree': [0.3, 0.7, 1.0]
}

# RandomizedSearchCV 설정
lgbm_search = RandomizedSearchCV(
    LGBMClassifier(random_state=42),
    param_distributions=lgbm_params,
    n_iter=10,
    scoring='f1_macro',
    cv=5,
    random_state=42
)

# 검색 수행
lgbm_search.fit(X_train, y_train_encoded)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Random Forest 하이퍼파라미터 검색 범위 설정
rf_params = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 5, 10],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt']
}

# RandomizedSearchCV 설정
rf_search = RandomizedSearchCV(
    RandomForestClassifier(random_state=42),
    param_distributions=rf_params,
    n_iter=10,
    scoring='f1_macro',
    cv=5,
    random_state=42
)

# 검색 수행
rf_search.fit(X_train, y_train_encoded)
```

```python
from sklearn.model_selection import cross_val_score, KFold

# 모델 인스턴스 생성
xgb_model = XGBClassifier(**xgb_search.best_params_)
lgbm_model = LGBMClassifier(**lgbm_search.best_params_)
rf_model = RandomForestClassifier(**rf_search.best_params_)

# k-fold 교차 검증 설정
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# 각 모델에 대해 k-fold 교차 검증 수행 및 평균 점수 계산
xgb_cv_scores = cross_val_score(xgb_model, X_train, y_train_encoded, cv=kf, scoring='f1_macro')
lgbm_cv_scores = cross_val_score(lgbm_model, X_train, y_train_encoded, cv=kf, scoring='f1_macro')
rf_cv_scores = cross_val_score(rf_model, X_train, y_train_encoded, cv=kf, scoring='f1_macro')

# 평균 F1 점수 출력
print(f"XGBoost Average F1 Score: {np.mean(xgb_cv_scores)}")
print(f"LightGBM Average F1 Score: {np.mean(lgbm_cv_scores)}")
print(f"Random Forest Average F1 Score: {np.mean(rf_cv_scores)}")
```
